```python
#  Importing Libraries 

import pandas as pd 
import seaborn as sns 
import numpy as np 

import matplotlib
import matplotlib.pyplot as plt 
plt.style.use('ggplot')
from matplotlib.pyplot import figure 
%matplotlib inline 
matplotlib.rcParams['figure.figsize']=(12,8) #Adjust the configuration of the plots we will create 


#  Reading the data  

df= pd.read_csv(r'C:\Users\Huawei\Documents\Python\Movie Industry\movies.csv')


#  Exploring Data 

df.head()
```

```python
# Checking for missing Data 

for col in df.columns: 
    pct_missing = int(np.mean (df[col].isnull()))
    print('{} - {}%'.format(col,pct_missing))
```

```python
# Data types of our columns 

df.dtypes
```

```python
# Replacing NaN for 0

df['budget']= df['budget'].fillna(0)
df['gross']=df['gross'].fillna(0)

df.head()
```

```python
# Changing data type of columns 

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')



```

```python
# Sorting Values by Gross Revenue 

df.sort_values(by=['gross'],inplace = False, ascending = False)
df

```

```python
# Dropping any Duplicates
df['name'].drop_duplicates()
```

```python
# Data Ready for Working 

"""

Hypothesis:
__________________________________________________

> The Gross Revenue is Highly Correlated With the 
Amount of Budget 


""" 

```

```python
# Scatter Budget vs Gross Revenue with matplotlib

plt.scatter(x= df['budget'], y = df['gross'])
plt.title('Budget vs Gross Revenue')
plt.xlabel('Budget Invested')
plt.ylabel('Gross Revenue')

plt.show
```

```python
# Plotting budget vs gross using seaborn 

sns.regplot(x='budget',y='gross',data=df,scatter_kws={"color":"red"},line_kws={"color":"blue"})
```

```python
# Evaluating Correlations (Pearson,Spearman,Kendall)
correlation_matrix = df.corr(method='pearson',numeric_only=True) #pearson
```

```python
df.corr(method='spearman',numeric_only=True) #kendall
```

```python
df.corr(method='kendall',numeric_only = True) #kendall
```

```python
# Deploying the Data in a Heat Map Using Seaborn 

sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')


plt.show()
```

```python
""" Only Numeric Values are shown in the Heatmap 
    so our Correlation Matrix isn't complete.
    Lets assign random values to the categories 
    that are shown as strings
"""
```

```python
#     Creating Numeric Representations 

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')


plt.show()
```

```python
df_numerized.corr()
```

```python
# Getting a Resume of every variable with their corresponding
# correlations with every other variable in our DataFrame
  

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs.to_string())
```

```python
sorted_pairs = corr_pairs.sort_values(ascending = False)

print(sorted_pairs.to_string())
```

```python
# Filtrating high Correlations (>= .5)
high_correlation = sorted_pairs[ (sorted_pairs) > 0.5 ] 

high_correlation 
```
